```python
import pandas as pd

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
print(df)
```

Task 1 — Categorical Encoding

Apply the correct encoding method to each categorical column:

Gender → Label Encoding as we can divide into 0 for male and 1 for female
City → One-Hot Encoding (drop one column to avoid the dummy variable trap)- We can add 2 columns as mumbai and delhi and add the flag to the coresponding rows
Size → Label Encoding with the correct order: Small=0, Medium=1, Large=2

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['gender_decoded']=le.fit_transform(df['Gender'])
df['size_decoded']=le.fit_transform(df['Size'])
df_encoded= pd.get_dummies(df, columns=['City'], drop_first=True)
print(df)
print(df_encoded)
```

Task 2 — Feature Scaling

```python
from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler

print("Before scaling:")
print(df[['Age', 'Fare']].describe())
std_scaler = StandardScaler()
rob_scaler = RobustScaler ()
df[['Age']] = std_scaler.fit_transform(df[['Age']])
df[['Fare']] = rob_scaler.fit_transform(df[['Fare']])
print(df[['Age', 'Fare']].describe())
```

We are using the standard scalar for the age as the data looks normal and no outlier between the values.It convert numeric columns in df to mean=0, std=1

For Fare column, We are having the outlier in the table .So,we are using the robust scalar to ignore the extreme value
